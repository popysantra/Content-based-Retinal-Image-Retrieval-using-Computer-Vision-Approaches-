```python
import time
import cv2
import numpy as np
from skimage import feature

# Function to extract LBP features from an image
def extract_lbp_features(image):
    if len(image.shape) > 2:  # Check if the image is not grayscale
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale

    radius = 3
    n_points = 8 * radius
    method = 'uniform'
    lbp_image = feature.local_binary_pattern(image, n_points, radius, method)
    hist, _ = np.histogram(lbp_image.ravel(), bins=np.arange(0, n_points + 3), range=(0, n_points + 2))
    hist = hist.astype("float")
    hist /= (hist.sum() + 1e-7)
    return hist

# Path to the image
image_path = 'C:/Users/arpit/Documents/mthd=-/Green_Channel_Images/000c1434d8d7.png'

# Read the image
image = cv2.imread(image_path)
start_time = time.time()
# Extract LBP features
lbp_features = extract_lbp_features(image)
end_time = time.time()

# Calculate the time taken
time_taken = end_time - start_time
# Print the features
print("LBP Features:")
print(lbp_features)
print(f"Time taken to extract LBP features: {time_taken:.6f} seconds")
```

```python
import cv2
import numpy as np
from skimage import feature
import time

# Function to extract LTP features from an image
def extract_ltp_features(image, threshold=5):
    if len(image.shape) > 2:  # Check if the image is not grayscale
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale

    radius = 3
    n_points = 8 * radius
    lbp = feature.local_binary_pattern(image, n_points, radius, method='default')
    
    # Applying the LTP transformation
    ltp_upper = np.where(image >= threshold, 1, 0)
    ltp_lower = np.where(image <= -threshold, -1, 0)
    
    ltp = ltp_upper + ltp_lower

    # Compute the histogram of the LTP
    hist, _ = np.histogram(ltp.ravel(), bins=np.arange(-1, 2), range=(-1, 1))
    hist = hist.astype("float")
    hist /= (hist.sum() + 1e-7)
    return hist

# Path to the image
image_path = 'C:/Users/arpit/Documents/mthd=-/Green_Channel_Images/000c1434d8d7.png'

# Read the image
image = cv2.imread(image_path)

# Measure time before feature extraction
start_time = time.time()

# Extract LTP features
ltp_features = extract_ltp_features(image)

# Measure time after feature extraction
end_time = time.time()

# Calculate the time taken
time_taken = end_time - start_time

# Print the features and time taken
print("LTP Features:")
print(ltp_features)
print(f"Time taken to extract LTP features: {time_taken:.6f} seconds")
```

```python
import cv2
import numpy as np
from skimage.feature import greycomatrix, greycoprops
import time

# Function to extract GLCM features from an image
def extract_glcm_features(image, distances=[1], angles=[0], levels=256, symmetric=True, normed=True):
    if len(image.shape) > 2:  # Check if the image is not grayscale
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale

    # Compute GLCM
    glcm = greycomatrix(image, distances=distances, angles=angles, levels=levels, symmetric=symmetric, normed=normed)

    # Extract GLCM properties
    contrast = greycoprops(glcm, 'contrast')[0, 0]
    dissimilarity = greycoprops(glcm, 'dissimilarity')[0, 0]
    homogeneity = greycoprops(glcm, 'homogeneity')[0, 0]
    energy = greycoprops(glcm, 'energy')[0, 0]
    correlation = greycoprops(glcm, 'correlation')[0, 0]
    asm = greycoprops(glcm, 'ASM')[0, 0]

    # Return the features as a numpy array
    features = np.array([contrast, dissimilarity, homogeneity, energy, correlation, asm])
    return features

# Path to the image
image_path = 'C:/Users/arpit/Documents/mthd=-/Green_Channel_Images/000c1434d8d7.png'

# Read the image
image = cv2.imread(image_path)

# Measure time before feature extraction
start_time = time.time()

# Extract GLCM features
glcm_features = extract_glcm_features(image)

# Measure time after feature extraction
end_time = time.time()

# Calculate the time taken
time_taken = end_time - start_time

# Print the features and time taken
print("GLCM Features:")
print(glcm_features)
print(f"Time taken to extract GLCM features: {time_taken:.6f} seconds")
```
